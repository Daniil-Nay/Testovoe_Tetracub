$Рассмотрим,\;какие\;результаты\;дают\;модели\;регрессии\;и\;оценим\;с\;помощью\;метрики\;MAE$
$Предскажем\;значения\;на\;2022\;и\;сравним\;истинные\;с\;предсказанными$<br>

```python
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
def training(model,X,y):
    X_train,x_test,Y_train,y_test = train_test_split(X,y,test_size=0.2,train_size=0.8,random_state=42)
    model.fit(X_train,
             Y_train)
    mse = mean_squared_error(y_test,model.predict(x_test))
    y_pred = model.predict(x_test)
    plt.scatter(range(len(y_test)),y_test,color='red')
    plt.scatter(range(len(y_test)),y_pred,color='blue')
    print("MSE",mse)
    plt.legend(['Истинные значения', 'Предсказанные значения'])
    plt.show()
```

```python


model = LinearRegression()
X,y = final_df.drop(columns=['2022','Country_']), final_df['2022']
training(model,X,y)
```

```python
model = GradientBoostingRegressor()
training(model,X,y)
```

```python
model = RandomForestRegressor()
training(model,X,y)
```

```python
model = xgb.XGBRegressor()
training(model,X,y)
```

```python
model = SVR()
training(model,X,y)
```

$Видно,\;что\;более\;точный\;результат\;предоставляет\;модель\;SVR$<br>
$Далее\;сравним\;предсказание\;SVR\;и\;ARIMA\;обучив\;только\;на\;данных\;с\;1961\;по\;2021$<br>

```python
from pmdarima import auto_arima

adf = final_df.copy(deep=True)
adf.set_index('Country_',inplace=True)
adf = adf.T
adf.index = pd.to_datetime(adf.index, format='%Y')
predicted_values = []
true_values = []
total_mse = 0
for country in adf.columns:
    country_data = adf[country]
    train_data = country_data[:-1]  
    test_data = country_data[-1:]   
    model = auto_arima(train_data,seasonal=False)
    forecast = model.predict(n_periods=1)
    print(f'predskazanie for {country}:',forecast,'istinnoe',test_data)
    mse = mean_squared_error(test_data,forecast)
    print('MSE',mse)
    predicted_values.append(forecast[0])
    true_values.append(test_data.values[0])
    total_mse +=mse
plt.scatter(range(len(true_values)), true_values, color='red', label='Истинные значения')
plt.scatter(range(len(predicted_values)), predicted_values, color='blue', label='Предсказанные значения')
plt.title("Сравнение истинных и предсказанных данных для всех стран в 2022 году")
plt.ylabel('Температура')
plt.legend()
plt.grid(True)
plt.show()
print("total_mse",total_mse)
```

```python
from sklearn.svm import SVR
import numpy as np

forecast_per_country_2022 = [] 
mse_total = 0
for country in final_df['Country_'].unique():
    country_data = final_df[final_df['Country_'] == country]
    X_train = np.array(range(1961, 2022)).reshape(-1, 1)  
    Y_train = country_data.loc[:, '1961':'2021'].values.flatten()  
    
    svr_model = SVR(kernel='rbf')  
    svr_model.fit(X_train, Y_train)
    
    forecast_2022 = svr_model.predict([[2022]])
    forecast_per_country_2022.append((country, forecast_2022))

for country, forecast_2022 in forecast_per_country_2022:
    true_temp_2022 = final_df[final_df['Country_'] == country]['2022'].values[0] 
    print(f"Сравнение предсказанного и истинного значения для {country} в 2022 году:")
    print(f"Предсказанное значение: {forecast_2022}, Истинное значение: {true_temp_2022}")
    mse = mean_squared_error([true_temp_2022], [forecast_2022])
    print(f"MSE для {country} в 2022 году:", mse)
    mse_total += mse

    plt.scatter([country], [true_temp_2022], color='red')
    plt.scatter([country], [forecast_2022], color='blue')

print("total_mse",mse_total)
plt.xticks([])
plt.title("Сравнение истинных и предсказанных данных для всех стран в 2022 году")
plt.ylabel('Температура')
plt.grid(False)
plt.show()
```
